**Stable-Beluga-13**

```python
from langchain.llms.huggingface_pipeline import HuggingFacePipeline
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from transformers import BitsAndBytesConfig
```

```python
nf4_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16
)

model_id = "stabilityai/StableBeluga-13B"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=nf4_config,
    device_map="auto"
)

model.tie_weights()

pipe = pipeline("text-generation",
                model=model,
                tokenizer=tokenizer,
                max_new_tokens=256
                )

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
res = llm("Who is the PM of India?")
print(res)
```

```python
documents=SimpleDirectoryReader("/content/data").load_data()
documents
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index import ServiceContext
from llama_index.embeddings import LangchainEmbedding

embed_model=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
service_context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
service_context
```

```python
index=VectorStoreIndex.from_documents(documents,service_context=service_context)
```

```python
index
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("What is Fundamental duties?")
```

```python
print(response)
```

```python
queries = ['What is the significance of the Preamble in the Constitution of India, and what are its key components?',
           'Explain the concept of Fundamental Rights as enshrined in the Constitution of India. Discuss any two fundamental rights and their implications.',
           'Define the concept of Directive Principles of State Policy. How do they differ from Fundamental Rights? Provide examples of Directive Principles.',
           'Discuss the procedure for amendment of the Constitution of India. What is the significance of the Basic Structure Doctrine in the context of constitutional amendments?',
           'Explain the composition and functions of the President of India as outlined in the Constitution.',
           'Describe the powers and functions of the Supreme Court of India as provided in the Constitution.',
           'Discuss the federal structure of the Indian Constitution. Highlight the distribution of powers between the Union and State governments.',
           'What are the qualifications required to become a member of the Rajya Sabha (Council of States)? How are members of the Rajya Sabha elected?',
           'Explain the concept of judicial review as implied in the Indian Constitution. How does it act as a check on the legislative and executive branches?',
           'Discuss the significance of the 73rd and 74th Constitutional Amendments in the context of decentralization of power and governance in India.'
           ]
```

```python
QnA = {}
for i,query in enumerate(queries):
  res = query_engine.query(query)
  QnA[query] = res
  print(i,': ',res)

```

```python
for k,v in QnA.items():
  print(k,': ','\n',v,'\n\n')
```
